```python
#In the titles column there are "Reasons/Departments" specified before the title code. 
#These are EMS, Fire, and Traffic. Use .apply() with a custom lambda expression to create a new column called "Reason" that contains this string value.
For example, if the title column value is EMS: BACK PAINS/INJURY , the Reason column value would be EMS.
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
df['Reason'].head()
```

```python
#What is the most common Reason for a 911 call based off of this new column?
```

```python
df['Reason'].value_counts()
```

```python
#Now use seaborn to create a countplot of 911 calls by Reason.
```

```python
sns.countplot(x='Reason',data=df,palette='viridis')
```

```python
#What is the data type of the objects in the timeStamp column?
```

```python
type(df['timeStamp'].iloc[0])
```

```python
#Use pd.to_datetime to convert the column from strings to DateTime objects.
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
df['timeStamp'].head(1)
```

```python
#We can grab specific attributes from a Datetime object like -
```

```python
time = df['timeStamp'].iloc[0]
time.hour
```

```python
#Use .apply() Method, to create 3 new columns called Hour, Month, and Day of Week. 
#You will create these columns based off the timeStamp column.
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
df.head()
```

```python
#Day of Week is an integer 0-6. Use the .map() 
#with this dictionary to map the actual string names to the day of the week:
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dmap)
df.head()
```

```python
#Use seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column.
```

```python
sns.countplot(x='Day of Week',data=df,hue='Reason',palette='viridis')
# Relocation of the legends outside
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
```

```python
#Now do the same for Month:
```

```python
sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')
# Relocation of the legends outside
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
```

```python
#Create a new column called 'Date' that contains the date from the timeStamp column. 
#You'll need to use apply along with the .date() method.
```

```python
df['Date']=df['timeStamp'].apply(lambda t: t.date())
df.head(2)
```

```python
#Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.
```

```python
df1=df.groupby('Date').count()
df1
```

```python
df.groupby('Date').count()['lat'].plot()
plt.tight_layout()
```

```python
#Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['lat'].plot()
plt.title("Traffic")
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['lat'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['lat'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
#Let's try to create heatmaps with seaborn and our data. But, first we will need to restructure the 
#dataframe so that the columns become the Hours and the Index becomes the Day of the Week. 
#There are lots of ways to do this, but let's try to combine groupby 
#with an unstack method.
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack() ##Here, we can also use pivot method
dayHour.head()
```

```python
#Now create a HeatMap using this new DataFrame.
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour)
```

```python
#Now create a clustermap using this DataFrame.
```

```python
sns.clustermap(dayHour,cmap='viridis')
```

```python
#Now repeat these same plots and operations, for a DataFrame that shows the Month as the column.
```

```python
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth)
```

```python
sns.clustermap(dayMonth,cmap='viridis')
```
